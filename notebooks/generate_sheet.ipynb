```python
# --- CONFIGURATION DU "STAB" (ACCORD RYTHMIQUE) ---

ROOT = "C"
CHORD_TYPE = "m9"  # C - Eb - G - Bb - D
TITLE = "The 'House Stab' (Cm9)"
DESC = "L'accord jou en 'Noires' (sur chaque Kick). \nNote la densit au piano (5 notes) et la position aigu  la guitare pour percer le mix."

# 1. PIANO : Voicing dense
# Note : m9 ajoute la 9me (R/D)  l'accord de m7.
# C(0) + Eb(3) + G(7) + Bb(10) + D(14)
piano_data = calculate_piano_indices(ROOT, CHORD_TYPE)

# 2. GUITARE : Position Funk "Hendrix/Frusciante"
# Position en Do (Case 8) sur la corde de Mi grave (souvent touffe) ou La.
# On va utiliser une forme sans racine (Rootless) ou Shell souvent utilise en rythmique.
# Forme : x - x - 8(Bb) - 8(Eb) - 8(G) - 10(D) -> Bb Major triad over C bass implied.
# C'est un voicing trs "Smooth".
guitar_data = [
    (0, -1),  # Mute
    (1, -1),  # Mute (Le bassiste joue le Do)
    (2, 8),  # Bb (Septime mineure) - Index barre
    (3, 8),  # Eb (Tierce mineure) - Index barre
    (4, 8),  # G (Quinte) - Index barre
    (5, 10),  # D (Neuvime) - Petit doigt (Le "Secret" de la couleur)
]

# 3. GNRATION
draw_sheet(
    title=TITLE,
    description=DESC,
    piano_indices=piano_data,
    guitar_states=guitar_data,
    save_path="../output/Cm9_House_Stab.png",
)
```
